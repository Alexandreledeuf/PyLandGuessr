```python
import pygame
from pygame.locals import *
import sys
from geopy.geocoders import Nominatim
from PIL import Image
import random
import os
import time
import cv2
import geopy.distance
import numpy as np


def resizeImage(weight, height):
    #resize image
    image = pygame.image.load(r'images/WorldMap.jpg')
    image = pygame.transform.scale(image, (weight, height))
    rect = image.get_rect()
    rect = rect.move((0, 0))
    return rect,image

def getLat(Y,coef_Y_positive,coef_Y_negative):
    if(Y> 0):
        if(Y >= 0 and  Y <=76):
            LAT = Y/coef_Y_positive[0]
        elif(Y>77 and Y <=149):
            LAT = Y/coef_Y_positive[1]
        elif(Y>150 and Y <=211):
            LAT = Y/coef_Y_positive[2]
        elif(Y>212 and Y <=263):
            LAT = Y/coef_Y_positive[3]
        elif(Y>264 and Y <=288):
            LAT = Y/coef_Y_positive[4]
        elif(Y>289 and Y <=300):
            LAT = Y/coef_Y_positive[5]
        else:
            LAT = 0
    else:
        if(Y > -84 and  Y <= 0):
            LAT = Y/coef_Y_negative[0]
        elif(Y > -154 and Y <= -85):
            LAT = Y/coef_Y_negative[1]
        elif(Y > -215 and Y <= -155):
            LAT = Y/coef_Y_negative[2]
        elif(Y > -265 and Y <= -216):
            LAT = Y/coef_Y_negative[3]
        elif(Y >= -300 and Y <= -266):
            LAT = Y/coef_Y_negative[4]
        else:
            LAT = 0
    return LAT

def getLon(X,coef_X):
    return X/coef_X


def getCountry(geolocalisation):
    geolocator = Nominatim(user_agent="http")
    location = geolocator.reverse(geolocalisation, language='en')
    if(location != None):
        adress = location.address.split(", ")
        return adress[len(adress)-1]
    else:
        coo = "Nothing"
        return coo
    
def getImage():
    path="./images/photos/"
    files=os.listdir(path)
    image =random.choice(files)
    image_path = str(path) + str(image)
    img = cv2.imread(image_path)
    cv2.imshow("test",img)
    image = image.replace(".jpg", "")
    image = image.lower()
    return image

def WrongAnswer(country,imageCountry, tryNumber,display,geolocalisation,geolocalisationImage, weight, height):
    
    distance = geopy.distance.geodesic(geolocalisation,geolocalisationImage).km
    distance = np.round(distance, 2)
    """ 
    r = 6371
    phi1 = np.radians(LAT)
    phi2 = np.radians(LatImage)
    delta_phi = np.radians(LAT - LatImage)
    delta_lambda = np.radians(LonImage - LON)
    a = np.sin(delta_phi / 2)**2 + np.cos(phi1) * np.cos(phi2) *   np.sin(delta_lambda / 2)**2
    res = r * (2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a)))
    test = np.round(res, 2)
    print(test)
    """

    falseReponse = "%s"% (country)
    Attempt = "%s attempts left"%(tryNumber)
    dist = "%s km away"%distance
    lenTexteFalse = len(falseReponse)
    lenAttempt = len(Attempt)
    lendist = len(dist)

    font = pygame.font.Font("freesansbold.ttf", 25)
    textFalse = font.render(falseReponse,True,(255,0,0))
    textAttempt = font.render(Attempt,True,(255,0,0))
    textdist = font.render(dist,True,(255,0,0))

    display.blit(textFalse,((weight/2)-5*lenTexteFalse,height/2))
    display.blit(textAttempt,((weight/2)-5*lenAttempt,(height/2)+25))
    display.blit(textdist,((weight/2)-5*lendist,(height/2)+50))
    pygame.display.update()
    time.sleep(3)
    
def gameOver(country,display,weight, height):
    falseReponse = "%s"% (country)
    lenTexteFalse = len(falseReponse)

    Loose = "You loose !"
    lenLoose = len(Loose)

    font = pygame.font.Font("freesansbold.ttf", 25)
    textFalse = font.render(falseReponse,True,(255,0,0))
    textLoose = font.render(Loose,True,(255,0,0))

    display.blit(textFalse,((weight/2)-5*lenTexteFalse,height/2))
    display.blit(textLoose,((weight/2)-5*lenLoose,(height/2)+25))
    pygame.display.update()
    time.sleep(2)
    cv2.destroyAllWindows()
    pygame.quit()
    sys.exit()

def GoodAnswer(country,display,weight,height):
    goodReponse = "%s"% (country)
    Victory = "You find the country !!!"
    lenTexteTrue = len(goodReponse)
    lenVictory = len(Victory)
    font = pygame.font.Font("freesansbold.ttf", 25)
    textGood = font.render(goodReponse,True,(0,255,0))
    textVictory = font.render(Victory,True,(0,255,0))

    display.blit(textGood,((weight/2)-5*lenTexteTrue,height/2))
    display.blit(textVictory,((weight/2)-5*lenVictory,(height/2)+25))
    pygame.display.update()
    time.sleep(2)

def main():
    pygame.init()
    weight = 1200
    height = 600
    display = pygame.display.set_mode((weight, height))
    FPS_CLOCK = pygame.time.Clock()
    rect,image = resizeImage(weight, height)
    
    coef_Y_positive = [5.8,5.5,4.7,4.5,3.8,3.5]
    coef_Y_negative = [5.78,5.7,5.0,4.6,3.5]
    coef_X = 3.35
    
    imageCountry = getImage()
    tryNumber = 3
    while True:
        display.blit(image, rect)
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                sys.exit()

            if event.type == pygame.MOUSEBUTTONDOWN:
                mouse_presses = pygame.mouse.get_pressed()
                if mouse_presses[0]:
                    coordinates = pygame.mouse.get_pos()
                    coordinates = list(coordinates)
                    X = coordinates[0]
                    Y = coordinates[1]
                    X = (X-(weight/2))
                    Y = -(Y-(height/2))
                    
                    LAT = getLat(Y,coef_Y_positive,coef_Y_negative)
                    LON = getLon(X,coef_X)
                    geolocalisation = str(LAT)+","+str(LON)
                    
                    country = getCountry(geolocalisation)
                    country = country.lower()
                    
                    # calling the Nominatim tool
                    loc = Nominatim(user_agent="GetLoc")

                    # entering the location name
                    getLocImage = loc.geocode(imageCountry)
                    LatImage = getLocImage.latitude
                    LonImage = getLocImage.longitude
                    geolocalisationImage = (str(LatImage) +","+ str(LonImage))
                    
                    if(country!="nothing"):
                        if(imageCountry==country):                 

                            GoodAnswer(country,display,weight,height)
                            
                            cv2.destroyAllWindows()
                            pygame.quit()
                            sys.exit()
                        else:
                            tryNumber -= 1
                            if(tryNumber>0):
                                WrongAnswer(country,imageCountry, tryNumber,display,geolocalisation,geolocalisationImage ,weight, height)
                            else:
                                gameOver(country,display,weight, height)

                    else:
                        badLocation = "Nothing was find here"
                        lenbadLocation = len(badLocation)

                        font = pygame.font.Font("freesansbold.ttf", 25)
                        textNothing = font.render(badLocation,True,(255,0,0))

                        display.blit(textNothing,((weight/2)-5*lenbadLocation,height/2))
                        pygame.display.update()
                        time.sleep(2)
                        
        pygame.display.update()
        FPS_CLOCK.tick(30)
        
if __name__ == "__main__":
    main()
```
